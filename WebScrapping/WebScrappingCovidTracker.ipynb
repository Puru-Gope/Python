```python
from bs4 import BeautifulSoup
import requests

r=requests.get("https://www.mygov.in/covid-19")
#print(r.ok)
data=r.content
soup=BeautifulSoup(data,"html.parser")
#print(soup.prettify())
level=soup.find("div",class_="information_row")
active=level.find("div",class_="iblock active-case").find("span",class_="icount").text
discharge=level.find("div",class_="iblock discharge").find("span",class_="icount").text
death=level.find("div",class_="iblock death_case").find("span",class_="icount").text
#print(level.prettify())
```

```python
print(active + ":" +death+ ":" +discharge)
```

```python
table_db=soup.find("div",class_="state_record")
print(table_db)
```

```python
#for head in table_db.find("thead").find_all("tr"):
   # print(head.text,end=" ")
MyList=[]

for row in table_db.find_all("tr"):
    MyRow=[]
    for cell in row.find_all("td"):
       # print(cell.text,end=",")
        MyRow.append(cell.text)
    MyList.append(MyRow)
del MyList[0]   
MyList
```

```python
import pandas

df=pandas.DataFrame(MyList,columns=["State/UTs","Confirmed","Active","Recovered","Deceased"])

df
df.to_csv('Covid_Table_23_06_20.csv')
```
